```python
scores=[]
choice=None

while choice !="0":
    print \
    """ 
    High Scores Keeper
    0-Exit
    1-Show High Scores
    2-Add a Score
    """  
    
    choice=raw_input("Choice: ")
    if choice== "0":
        print "Good-bye."
    
    elif choice == "1":
        print "High Scores"
        for entry in scores:
            score,name=entry
            print name, "\t", score
    elif choice =="2":
        name=raw_input("What is the player's name?: ")
        score=int(raw_input("What score did you get?:"))
        entry=(score,name)
        scores.append(entry)
        scores.sort()
        scores.reverse()
        scores=scores[:5]
    else:
        print "Sorry, but", choice, "isn't a valid choice."  
raw_input("\n\nPress the enter key to exit.")
```

```python
import random
MAX_WRONG=10
WORDS=("OVERUSED", "CLAM", "GUAM", "PUCK", "TAFFETA")
word=random.choice(WORDS)
so_far = "-" * len(word)
wrong=0
used=[]
print "Welcome to Hangman.  Good luck!"
while wrong <MAX_WRONG and so_far!=word:
    print "So far you have used", wrong, "guesses"
    print "\n You've used the following letters: \n", used
    print "\n So far, the word is: \n", so_far
    guess=raw_input("\n\nEnter your guess: ")
    guess=guess.upper()
    
    while guess in used:
        print "You've already guessed the letter:", guess
        guess=raw_input("Enter your guess: ")
        guess=guess.upper()
    used.append(guess)  
    
    if guess in word:
        print "\n Yes!", guess, "is in the word!"
        new="" 
        for i in range(len(word)):
            if guess==word[i]:
                new+=guess
            else:
                new+=so_far[i]
        so_far=new
    else:
        print "\nSorry,", guess, "isn't in the word."
        wrong +=1
if wrong ==MAX_WRONG:
    print "You have used your maximum number of guesses"
else:
    print"\nYou guessed it!"
print "\nThe word was", word
raw_input("\n\n Press the enter key to exit.")
```

```python
# Band Song Program
#Given a band, print the name of the bands top song
band={"CAGE THE ELEPHANT"  :  "SHAKE ME DOWN" ,}
choice=None
while choice != "0":
    print \
    """  
    0-Quit
    1-Look up a band's top song
    2-Add a band's top song
    3-Redefine a band's top song
    4-Delete a band
    """  
    choice=raw_input("Choice: ")
    if choice == "0":
        print "Bye."
    elif choice == "1":
        search=raw_input("What band do you want to search for?")
        search=search.upper()
        print search
        if search in band:
            top_song=band[search]
            print "\n", search,"'s top song is", top_song
        else:
            print "\nSorry, I don't know", search
    elif choice == "2":
        new_band=raw_input("What band do you want me to add?: ")
        new_band=new_band.upper()
        if new_band not in band:
            top_song=raw_input("What is their top song?")
            band[new_band]=top_song
            print "\n", new_band, "has been added."
        else:
            print "\nThat band already exists!  Try redefining their top song."
    elif choice == "3":
        changed_band=raw_input("What band do you want me to change? ")
        changed_band=changed_band.upper()
        if changed_band in band:
            new_song=raw_input("What is their new top song?")
            band[changed_band]=new_song
            print "\n", changed_band, "has been changed."
        else:
            print "\n", changed_band, "does not exist.  Consider adding it with choice 2."
    elif choice == "4":
        del_band=raw_input("What band do you want me to delete?: ")
        del_band=del_band.upper()
        if del_band in band:
            del band[del_band]
            print"\nOkay, I deleted", del_band
        else:
            print "\nIcan't do that!", del_band, "does not exist."
    else:
        print"\n Sorry, but", choice, "is not valid."
raw_input("\n\nPress the enter key to exit.")
```
